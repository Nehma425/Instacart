```python
# Creating a histogram on 'order_hour_of_day' column using 24 bins for 24 hours in a day

hist = ords_prods_custs_merge['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram shows that business is slow in the early hours, then peaks from 10 am to 3 pm, and then starts to decline.

```python
# Creating bar chart from 'loyalty_flag' column

bar = ords_prods_custs_merge['loyalty_flag'].value_counts().sort_index().plot.bar()
```

Most customers are considered "Regular" followed by "loyal" and finally "new"
